```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from PIL import Image
```

```python
import os
image_file = os.path.join("../test_images", "truck.jpg")
if not os.path.exists(image_file):
    raise FileNotFoundError(f"Image file not found: {image_file}")
image = cv.imread(image_file)
if image is None:
    raise ValueError("Failed to read the image. Ensure the file path is correct.")
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title("Loaded Image")
plt.axis("off")
plt.show()
```

```python
# Denoise Image

# using cv.fastNlMeansDenoisingColored
denoised_image = cv.fastNlMeansDenoisingColored(image, None, 10, 10, 7, 21)
plt.imshow(cv.cvtColor(denoised_image, cv.COLOR_BGR2RGB))
plt.title("Denoised Image")
```

```python
# sharpen Image
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
sharpened_image = cv.filter2D(denoised_image, -1, kernel)
plt.imshow(cv.cvtColor(sharpened_image, cv.COLOR_BGR2RGB))
```

```python
# save image
output_file = os.path.join("../test_images", "truck_processed.jpg")
cv.imwrite(output_file, sharpened_image)
```
